### Representacion grafica de los Datos

si a la funci칩n de trazado se le da una matriz de datos, usara como coordenadas en el eje vertical, y utilizara el indice de cada punto d datos en el array como la coordenada horizontal.

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.show()
```

Pueden modificarse los ejes de las figuras para que no se mira tan ajustada

Se sigue el mismo procedimento para pintar una funcion matematica.

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2

plt.plot(x,y)
plt.show()
```

Tambien pueden modificsrse el estilo de la grafica para que contenga mas informacion.

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Juanita")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

Pueden superponerse graficos y cambiar el estilo de las lineas

```python
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1

plt.plot(x, y, "b--", x, y2, "y")
plt.show()
```

```python
# Separado en diferentes lineas las funciones
x= np.linspace(-2,2,500)
y = x**2
y2 = x + 1

plt.title("Square Function")
plt.plot(x, y, "b--")
plt.plot(x, y2, "g")
plt.show()
```

```python
# Para poder diferenciar entre las funciones siem9pre es recomendble a침adir una leyenda
# Separado en diferentes lineas las funciones
x= np.linspace(-2,2,500)
y = x**2
y2 = x + 1

plt.title("Square Function")
plt.plot(x, y, "b--", label = "x**2")
plt.plot(x, y2, "g", label = "x+1")
plt.legend("best")
plt.show()
```

#### Tambien puede Crearse dos graficas que no se superpongan estas graficas se organizan en un grid y se denominan subplots 

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x +1

plt.subplot(1, 2, 1)  # 1 rows, 2 columns, 1st subpot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2)  # 1 rows, 2 columns 2do subplot
plt.plot(x, y2, 'g')

plt.show()
```
